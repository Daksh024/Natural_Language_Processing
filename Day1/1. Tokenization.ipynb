```python
sents = "Hello friends! How are you? Welcome to Python Programming. "
```

```python
from nltk.tokenize import word_tokenize
```

```python
word_tokenize(sents)
```

```python
sent = 'They told that their ages are 23, 25 and 29 respectively.'
```

```python
# 1. find the avg of ages
# 2. total no of counts starts with "t"/"T".
# 3. count total no of punctuations symbols.
# 4. print the word in capital letters which ends with "d".
# 5. merge all the words and print (No punctuation)
```

```python
words = word_tokenize(sent)
nums = [int(word) for word in words if word.isdigit()]
```

```python
sum(nums)/len(nums)
```

```python
s = 0
for word in words:
    if word.startswith("t") |  word.startswith("T"):
        s +=1

print(f"total no of counts starts with 't'/'T' is {s}.")
```

```python
words
```

```python
c = 0
for word in words:
    if not word.isalnum():
        c += 1

print(f"count total no of punctuations symbols is :{c}")
```

```python
char = []
for word in words:
    if word.endswith('d'):
        char.append(word.upper())

print(f"print the word in capital letters which ends with 'd' :{char}")
```

```python
char
```

```python
ls = []

for word in words:
    if word.isalnum():
        ls.append(word)

''.join(ls)
```

```python
''.join(word for word in words if word.isalnum())
```

### Multi Word Expression Tokenizer

```python
sent = ' rassam rice is good at the place that mentioned by srujana mam.'
```

```python
from nltk.tokenize import MWETokenizer
mwet = MWETokenizer(separator='-')
mwet.add_mwe(('rassam','rice'))
# mwet.add_mwe()
mwet.tokenize(word_tokenize(sent))
```
