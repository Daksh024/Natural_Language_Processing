```python
sents
```

```python
from nltk.tokenize import word_tokenize
```

```python
tokens = word_tokenize(sents)
```

```python
tokens
```

```python
from nltk import FreqDist
```

```python
freq = FreqDist(tokens)
```

```python
freq
```

```python
freq.plot(20,cumulative=False)
```

### Clean list without any punctuations

```python
cln_tok = [word for word in tokens if word.isalnum()]

## another way
import string
string.punctuation
# cln_tok = [word for word in tokens if word not in string.punctuation]
```

```python
cln_tok
```

```python
freq = FreqDist(cln_tok)
freq.plot(20,cumulative=False)
```

### Removing stopwords

```python
stp = [word for word in cln_tok if word not in swords]
```

```python
stp
```

```python
freq = FreqDist(stp)
freq.plot(20,cumulative=False)
```
